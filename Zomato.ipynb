```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step2 - Create the data frame

```python
df = pd.read_csv("Zomato data .csv")
```

```python
df
```

# Convert the data type of column rate

```python
def handleRate(value):
    value = str(value).split('/')# split the rate values after ' /' as 41.5/5
    value = value[0] # starting from 1 index
    return float(value) # taking the values as flaot 41.5 after splitting

# Now applying above user defined function in the dataframe column rate
df['rate'] = df['rate'].apply(handleRate)
df.head()
```

```python
df.info()
```

# Type of Restaurant
Majority orders the food from which resataurant

```python
df.head()
```

```python
# using seaborn library to crearte the plot
sns.countplot(x =df['listed_in(type)'])# countplot displays the count of observations
plt.xlabel('Type of restaurant')
plt.show()
```

# conclusion - majority of the restaurant falls in dining category

```python
df.head()
```

# Analyzing the votes

```python
# Creating plot usimg matplotlib library

grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})# coverting result in dataframe


# creating line graph in matplotlib
plt.plot(result ,c='green',marker='o')
plt.xlabel('Type of restaurant' , c='red' ,size=20)
plt.ylabel('votes' , c='red' ,size=20)

# Saving the plot as image
plt.savefig('restaurant_votes.png', dpi=300,
           bbox_inches='tight')
plt.show()
```

# conclusion - dining restaurants has received maximum votes

```python
plt.figure(figsize = (4,4))
sns.boxplot(x ='online_order' ,y ='rate' ,data=df)
plt.show()
```

```python
df.head()
```

```python
pivot_table = df.pivot_table(index ='listed_in(type)', columns ='online_order' ,
                            aggfunc = 'size' ,fill_value=0)

sns.heatmap(pivot_table, annot = True , cmap ='YlGnBu' , fmt='d')# cmap set the color 

plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed in(Type)')
plt.show()
```

# conclusion - Dining restaurants primarily accepts offline orders,
# whereas cafes primarily receives online orders.This suggests that clients 
# prefer orders in person at restaurants,  but prefer online ordering at cafes.
